````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("previsao_de_renda.csv")
df.head()
````

````python
ax = sns.countplot(x='data_ref', hue='posse_de_veiculo', data=df)

tick_labs = df['data_ref'].unique()

tick = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````

As alterações perceptíveis mais significativas encontram-se nas variáveis "posse_de_veiculo", "educacao" e "tipo_renda". Na primeira, observa-se que houve um aumento na diferença de renda média entre os clientes que possuiam algum tipo de veiculo em relação aos que não tinham durante o mês de setembro de 2015; em todos os períodos restantes, apesar da renda média dos possuidores de veiculo ser superior ao dos que não têm, há encontro entre as barras de erros de cada categoria. No segundo caso houve um pico na renda média dos clientes com pós-graduação em setembro de 2015, sendo muito superior a todos os outros tipos de educação; além disso, a mesma renda média também foi superior às outras entre os meses de abril e junho de 2015; em todos os outros casos, não havia variação entre as rendas de cada nível de educação. Na última, nota-se que houve uma grande queda na renda média dos bolsistas entre fevereiro e março de 2015, desaparecendo do gráfico após isso; ela reaparece em dezembro do mesmo ano, ainda sendo menor do que os outros tipos de renda. 
Aparentemente, aúnica variável qualitativa que poderia ser utilizada em uma modelagem da renda dos clientes seria a "posse_de_veiculo", uma vez que esta é a única que mostra diferença entre as rendas médias dos clientes. A variável "posse_de_imovel" também apresenta um certo potencial para ser utilizada, mas não tão alto quanto a primeira, devido à pequena diferença entre as médias.
