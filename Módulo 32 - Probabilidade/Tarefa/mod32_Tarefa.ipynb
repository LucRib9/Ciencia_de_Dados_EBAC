1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

> Para resolver esta primeira questão, é necessário usar a Função de Distribuição Acumulativa (**F(X)**), responsável por medir a probabilidade de uma variável aleatória ser menor do que um determinado valor (**P(X < x)**):
>
> $$ F(X < x) = P(X < x)  $$
>
> Considerando que queremos saber a probabilidade do pacote selecionado ter mais do que 15,2 kg e que, em uma distribuição Normal, o intervalo de valores é infinito, precisamos fazer o cálculo:
>
> $$ P(X > 15,2) = F(\infty) - F(15,2)$$ 
>
> O cálculo desta probabilidade encontra-se abaixo.

```python
# Média "mu" e Desvio Padrão "sigma"
mu = 15
sigma = 0.1

# Distribuição Normal "N" com a Média de Desvio Padrão informados
N = stats.norm(loc=mu, scale=sigma)

# Valores mínimo "low" e máximo "high"
low = 15.2 
high = np.inf

# Probabilidade do peso maior que 15.2 kg
prob = N.cdf(high) - N.cdf(low)

print(f'A probabilidade para valores maiores que {low:.1f} kg é de {prob:.2%}'.replace('.',',')+'.')
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

> Para calcular o percentil de ordem 95%, precisamos usar a Função Ponto Percentil, sendo esta a inversa da CDF. Ou seja, ela devolve o valor máximo da variável aleatória, para que uma seleção de qualquer valor menor do que o máximo tenha probabilidade na ordem da informada. 
> Na célula abaixo encontra-se o cálculo do peso dos pacotes.

```python
# Calcula a Função Ponto Percentil para 95% para o peso dos pacotes.
ppf = N.ppf(.95)

# Imprime o peso dos pacotes que obedece à PPF(95%)
print(f'Há 95% de chances de se obter pacotes com peso inferior a {ppf:.2f} kg'.replace('.',',') + '.')
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

> Para que sejam calculadas as variáveis aleatórias para cada valor do objeto **Y**, vamos criar dois vetores: **Y1** e **Y2**, sendo que o primeiro tem distribuição Normal padrão, enquanto que o segundo, distribuição t-student. 

```python
# Vetor com os valores do objeto Y (valores entre -6 e 6)
Y=np.linspace(-6, 6, 200)

# Parâmetros da distribuição Normal do vetor Y1 (média, variância e Desvio Padrão)
mu_Y1 = 0
var_Y1 = 1
sigma_Y1 = math.sqrt(var_Y1)

# Vetor Y1 com os valores da Função de Distribuição Acumulativa para todo o Y
Y1 = stats.norm.cdf(Y, mu_Y1, sigma_Y1) 

# Imprime o vetor Y1
Y1
```

```python
# Cria o vetor Y2 com 5 graus de liberdade, calculando a CDF para cada valor de Y  
Y2 = stats.t.cdf(Y, 5) 

# Imprime o vetor Y2
Y2
```

4. Faça uma visualização gráfica dos dados do item anterior.

> Agora que os vetores **Y1** e **Y2** estão prontos, vamos fazer o gráfico de seus valores. Para compará-los, vamos imprimir ambos no mesmo gráfico.

```python
# Cria uma figura com largura = 10 e altura = 4 para o gráfico
fig, ax = plt.subplots(figsize=(10,4))

# Plota o vetor Y1 em função de Y, rotulado por "Y1 - Normal"
ax.plot(Y, Y1, label='Y1 - Normal')

# Plota o vetor Y2 em função de Y, rotulado por "Y2 - t(5)"
ax.plot(Y, Y2, label='Y2 - t(5)')

# Insere a legenda (rótulo) no gráfico
ax.legend()
```

> O gráfico mostra que valores da CDF em ambas as distribuições são muito próximos um do outro, sendo que a t-student apresenta valores superiores à distribuição Normal para Y < 0, e o inverso para Y > 0.
