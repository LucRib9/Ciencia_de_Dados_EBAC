#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
%matplotlib notebook
plt.ion()
```

```python
# Carregamento dos dados
sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
# APGAR1
plt.close('all')
plt.rc('figure', figsize=(6, 5))
apgar1 = sns.countplot(x='APGAR1', data=sinasc, color='b', alpha=0.8)
apgar1.set(title='Apgar1', xlabel=None, ylabel='Contagem')
plt.show()
```

```python
# APGAR5
plt.close('all')
plt.rc('figure', figsize=(6, 5))
apgar5 = sns.countplot(x='APGAR5', data=sinasc, color='r', alpha=0.8)
apgar5.set(title='Apgar5', xlabel=None, ylabel='Contagem')
plt.show()
```

```python
# SEXO
plt.close('all')
plt.rc('figure', figsize=(6, 4))
sexo = sns.countplot(x='SEXO', data=sinasc, alpha=0.8)
sexo.set(title='Sexo da Criança', xlabel=None, ylabel='Contagem')
plt.show()
```

```python
# GESTACAO
plt.close('all')
plt.rc('figure', figsize=(7, 6))
gestacao = sns.countplot(x='GESTACAO', data=sinasc, alpha=0.8)
gestacao.set(title='Período de Gestação (em Semanas)', xlabel=None, ylabel='Contagem')
semanas = ['< 22', '22 a 27', '28 a 31', '32 a 36', '37 a 41', '>= 42']
gestacao.set_xticklabels(semanas, rotation=30)
plt.show()
```

```python
# ESTCIVMAE
plt.close('all')
plt.rc('figure', figsize=(7, 6))
estados = ['Solteira', 'Casada', 'Viuva', 'Divorciada', 'Consensual', 'Ignorado']
estcivmae = sns.countplot(x='ESTCIVMAE', data=sinasc, alpha=0.8)
estcivmae.set(title='Estado Civil da Mãe', xlabel=None, xticklabels = estados, ylabel='Contagem')
plt.xticks(rotation=30)
plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(7, 7))
fig = plt.figure()
consultas_apgar5 = sns.barplot(x='CONSULTAS', y='APGAR5', data=sinasc)
n_consultas = ['0', '1 a 3', '4 a 6', '>= 7', 'Ignorado']
consultas_apgar5.set(xlabel='Nº de Consultas', xticklabels = n_consultas, ylabel='Apgar5')
plt.xticks(rotation=30)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 9))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax=axes[0,0], x='SEXO', data=sinasc, alpha=0.8).set(xlabel=None, title='Sexo da Criança', ylabel=None)
sns.countplot(ax=axes[0,1], x='GESTACAO', data=sinasc, alpha=0.8).set(xlabel=None, ylabel=None, 
                                                                      title='Período de Gestação (em Semanas)')
sns.countplot(ax=axes[1,0], x='ESTCIVMAE', data=sinasc, alpha=0.8).set(xlabel=None, title='Estado Civil da Mãe', ylabel=None)
sns.countplot(ax=axes[1,1], x='CONSULTAS', data=sinasc, alpha=0.8).set(xlabel=None, title='Nº de Consultas', ylabel=None)

semanas = ['< 22', '22 a 27', '28 a 31', '32 a 36', '37 a 41', '>= 42']
axes[0,1].set_xticklabels(semanas, rotation=30)

estados = ['Solteira', 'Casada', 'Viuva', 'Divorciada', 'Consensual', 'Ignorado']
axes[1,0].set_xticklabels(estados, rotation=30)

n_consultas = ['0', '1 a 3', '4 a 6', '>= 7', 'Ignorado']
axes[1,1].set_xticklabels(n_consultas, rotation=30)

plt.subplots_adjust(wspace=0.3, hspace=0.4)
plt.show()
```
